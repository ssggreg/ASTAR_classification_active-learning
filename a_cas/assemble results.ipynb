```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import torch.nn.functional as F
sns.set()
```

```python
def plot_mean_std(percentages, accuracies, label, log=False, color="blue"):
    mean = np.mean(accuracies,axis=0)
    std  = np.std(accuracies,axis=0)
   
    plt.plot(percentages,mean, label=label, color=color)
    plt.fill_between(percentages, mean + std, mean - std, facecolor=color, alpha=0.5)
    if log: plt.xscale('log')
    plt.legend()
```

```python
vogn_bald= np.load('vogn_results/vogn_bald.npy')
vogn_ent= np.load('vogn_results/vogn_ent.npy')
vogn_f= np.load('vogn_results/vogn_f.npy')
vogn_std= np.load('vogn_results/vogn_std.npy')
vogn_var= np.load('vogn_results/vogn_var.npy')
```

```python
mc_bald= np.load('mc_results/mc_bald.npy')
mc_ent= np.load('mc_results/mc_ent.npy')
mc_f= np.load('mc_results/mc_f.npy')
mc_std= np.load('mc_results/mc_std.npy')
mc_var= np.load('mc_results/mc_var.npy')
```

```python
batch_size_sample = np.ones(261).astype(int)
for i in range(100):
    batch_size_sample[i+110]=3
for i in range(60):
    batch_size_sample[i+200]=5
```

```python
nnn = np.ones(262)
for i in range(261):
    nnn[i+1] = nnn[i] + batch_size_sample[i]
```

```python
plt.clf()
plot_mean_std(np.arange(684)[1:102]/3420,random[:,:101],'random selection',color='orange')
plot_mean_std(nnn[:101]/3420,vogn_bald[:,:101],'Vogn_bald selection',color='blue')
plot_mean_std(nnn[:101]/3420,vogn_ent[:,:101],'Vogn_ent selection',color='r')
plot_mean_std(nnn[:101]/3420,vogn_f[:,:101],'Vogn_f selection',color='m')
plot_mean_std(nnn[:101]/3420,vogn_std[:,:101],'Vogn_std selection',color='y')
plot_mean_std(nnn[:101]/3420,vogn_var[:,:101],'Vogn_var selection',color='k')
plt.savefig('graph_vogn_02.png')
```
