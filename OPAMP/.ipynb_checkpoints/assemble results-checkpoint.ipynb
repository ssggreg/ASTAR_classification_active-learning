```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import torch.nn.functional as F
sns.set()
```

```python
def plot_mean_std(percentages, accuracies, label, log=False, color="blue"):
    mean = np.mean(accuracies,axis=0)
    std  = np.std(accuracies,axis=0)
   
    plt.plot(percentages,mean, label=label, color=color)
    plt.fill_between(percentages, mean + std, mean - std, facecolor=color, alpha=0.5)
    if log: plt.xscale('log')
    plt.legend()
```

```python
random = np.load("random_even_faster.npy")
```

```python
random_w = np.load("rararra.npy")
```

```python
vogn_even_fast = np.load("VOGN_accuracies_20_fast_even.npy")
```

```python
dropout_fast = np.load("dropout_accuracies_20_fast_even.npy")
```

```python
batch_size_sample = np.ones(261).astype(int)
for i in range(100):
    batch_size_sample[i+110]=3
for i in range(60):
    batch_size_sample[i+200]=5
```

```python
nnn = np.ones(262)
for i in range(261):
    nnn[i+1] = nnn[i] + batch_size_sample[i]
```

```python
plt.clf()
plot_mean_std(np.arange(684)/3420,random,'random selection',color='orange')
plot_mean_std(nnn/3420,vogn_even_fast,'VOGN selection',color='blue')
plot_mean_std(nnn/3420,dropout_fast,'Dropout selection',color='green')
plt.savefig('graph_1.png')
```

```python
plot_mean_std(nnn[:100]/3420,vogn_even_fast[:,:100],'VOGN selection',color='blue')
plot_mean_std(np.arange(100)/3420,random[:,:100],'random selection',color='orange')
plot_mean_std(nnn[:100]/3420,dropout_fast[:,:100],'Dropout selection',color='green')

plt.savefig('graph_2.png')
```
